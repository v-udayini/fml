```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
```

```python
plt.plot(x,x,label='linear')
plt.show()
```

```python
plt.plot(x,x,label='non-linear')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='darkblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='.')
plt.xlim(0.5, 4.5)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 8, 6, 4])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 8, 6, 4])

plt.barh(x, y)
plt.show()
```

```python
x = np.random.normal(150, 20, 300)

print(x)


plt.hist(x)
plt.show()
```

```python
y=np.array([35,25,15,10,5,1])
plt.pie(y)
plt.show()
```

```python
y=np.array([50,40,30,20,10,5,3,1])
mylabels=["upsc","army","cbi","gate","iaf","rrb","ib","nda"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["upsc", "cbi", "rrb", "iaf"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
